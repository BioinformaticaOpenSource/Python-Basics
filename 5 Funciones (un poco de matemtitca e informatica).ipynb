## 5.1 Funcion matematica

En casi todo fenómeno físico se observa que una cantidad depende de otra. Por ejemplo, la estatura depende de la edad, la temperatura depende de la fecha y el costo de enviar un paquete depende de su peso. 
Se usa el termino función para esta dependencia de una cantidad sobre otra, es decir se expresa lo siguiente:

    •	La altura es una función de la edad
    •	La temperatura es una función de la fecha 
    •	El costo de enviar por correo un paquete es una función del peso
    
**Definición de función** 

Una función “f” es una regla que se asigna a cada elemento de “x” en un conjunto A exactamente un elemento, llamado “f(x)” en un conjunto B. 


Matemáticamente una función es una operación que toma uno o más valores llamados argumentos y produce un valor denominado resultado —valor de la función para los argumentos dados—. Todos los lenguajes de programación tienen funciones incorporadas, intrínsecas o internas


***También podemos pensar en una función como una máquina para transformar variables***


## 5.2 Funcion en programacion

Cada función se evoca utilizando su nombre en una expresión con los argumentos actuales o reales encerrados
entre paréntesis.

    nombre_funcion (argumento1, argumento2,) 

Las funciones incorporadas al sistema se denominan funciones internas o intrínsecas

Las funciones definidas por el usuario, funciones externas. 

A una función no se le llama explícitamente, sino que se le invoca o referencia mediante un nombre y una listade parámetros actuales. El algoritmo o programa llama o invoca a la función con el nombre de esta última en una expresión seguida de una    lista de argumentos que deben coincidir en cantidad, tipo y orden con los de la función que fue definida. La función devuelve  un único valor.

***Asi se ve una pequeña funcion en Python***



## 5.3 Funciones internas o intrínsecas









```python
"""
Probando diferentes funciones internas de Python

Este prorgrma vuelve absoluto una variable numerica usnado la 
funcion abs()
"""
# Entrada

var_NumeroNeg = -12.2
val_Abs = None

# Proceso

# Se invoca a la funcion y le pasamos como argumento la variable var_NumeroNeg
# y el resultado de la funcion se almacena en la variable val_Abs

val_Abs = abs(var_NumeroNeg)

# Salida

print(val_Abs)
```

```python
"""
Probando diferentes funciones internas de Python

Este prorgrma vuelve un numero redondeado de una variable numerica usnado la 
funcion round()
"""
# Entrada

var_Numero = -12.96658
val_round = None

# Proceso

# Se invoca a la funcion y le pasamos como argumento la variable var_NumeroNeg
# y el resultado de la funcion se almacena en la variable val_Abs

val_round = round(var_Numero)

# Salida

print(val_round)
```

## 5.4  Librerias 

### ***Algunas definiciones cortas y sencillas para definir una librería***

Una librería es un una serie de funcionalidades compiladas y codificadas en un lenguaje de programación las cuales sirven de unos métodos o interfaces para realizar la funcionalidad que se requiere. No es un programa que se ejecute de manera autónoma, su función es servir a otros programas que la invoquen o mismamente a otras librerías informáticas.

https://bytecode.es/que-es-una-libreria-informatica/

Una librería es una colección de programas o subrutinas que pueden ser utilizadas por programas independientes. En programación, la abstracción consiste en aislar un elemento de su contexto convirtiéndolo en una "caja negra" donde lo importante es el "¿qué hace?" más que el "¿cómo lo hace?". La programación modular es un claro ejemplo de abstracción a partir de las funciones o módulos de una librería.

http://agrega.juntadeandalucia.es/repositorio/21012019/89/es-an_2019012112_9123621/33_libreras_y_reutilizacin_del_cdigo.html


A grandes rasgos y sin entrar en detalles, una librería es un grupo de procedimientos (funciones) escritos para ser utilizados o ser llamados (invocados) en diferentes partes del código con la finalidad de ahorrar tiempo, espacio y esfuerzo. Las librerías, don importantes para realizar tareas especificas y especializadas. 


### ***Ejemplos de librerias***

    as     ==> palabra reservada para indicar como se invoca la librería
